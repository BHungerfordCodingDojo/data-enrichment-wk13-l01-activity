1. Which customer made the fewest purchases (invoices)?

2. What is the most a customer has spent on a single purchase?

3. What country is the customer with the largest total bill from (from question 2)?

4. Retrieve the id, name, and # of playlists for tracks that appear on playlists more than 4 times?


```python
# 1. Which customer made the fewest purchases (invoices)?
# I used the paintbrush for formatting
q = """SELECT 
    i.CustomerID, COUNT(*) AS purchases, c.FirstName, c.LastName
FROM
    invoice AS i
        JOIN
    customer AS c ON i.CustomerID = c.CustomerID
GROUP BY CustomerID
ORDER BY Purchases
LIMIT 2"""

pd.read_sql(q, engine)

# Limit 2 to see if there was a tie
```

```python
# 2. Which customer made the fewest purchases (invoices)?
q = """SELECT 
    MAX(Total)
FROM
    Invoice"""

pd.read_sql(q, engine)
```

```python
# 3. What country is the customer with the largest total bill from (from question 2)?
q = """SELECT 
    MAX(Total), BillingCountry
FROM
    invoice"""

pd.read_sql(q, engine)
```

```python
# 4. Retrieve the id, name, and # of playlists for tracks that appear on playlists more than 4 times?
q = """SELECT 
    p.TrackID, COUNT(*) AS number_of_playlists, t.name
FROM
    playlisttrack AS p
        JOIN
    track AS t ON p.TrackID = t.TrackID
GROUP BY TrackID
HAVING number_of_playlists > 4"""

pd.read_sql(q, engine)
```
